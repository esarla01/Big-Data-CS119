# Exercise 1: Web Mining

```python
import requests
```

```python
from bs4 import BeautifulSoup
```

```python
def init_url():
  return 'https://aliceinwonderland.fandom.com/wiki/Alice_in_Wonderland_Wiki'
```

```python
# 1 - fetches the webpage using requests.get() **bold text**\\
# 2 - parses the HTML content using BeautifulSoup with "lxml" parser

def get_soup(url):
  response = requests.get(url)
  return BeautifulSoup(response.text, "lxml")
```

```python
def books():
    url = init_url()
    soup = get_soup(url)

    # Find the "Literature" section
    literature_section = soup.find("span", string="Literature")
    if not literature_section:
        return []

    # Find the nearest "Canon books" dropdown
    canon_books_section = literature_section.find_next("span", string="Canon books")
    if not canon_books_section:
        return []

    # Navigate to the list under Canon Books
    books_list = canon_books_section.find_next("ul")
    if not books_list:
        return []

    # Extract book titles and URLs
    books = []
    for li in books_list.find_all("li"):
        a_tag = li.find("a")
        if a_tag:
            book_name = a_tag.text.strip()
            books.append(book_name)

    return books

# canon_books = books()
# print("Canon Books:", canon_books)
```

```python
def poems():
    url = init_url()
    soup = get_soup(url)

    # Find the "Literature" section
    literature_section = soup.find("span", string="Literature")
    if not literature_section:
        return {}

    # Find the nearest "Canon poems" dropdown
    canon_poems_section = literature_section.find_next("span", string="Canon poems")
    if not canon_poems_section:
        return {}

    # Navigate to the list under Canon Poems
    poems_list = canon_poems_section.find_next("ul")
    if not poems_list:
        return {}

    # Extract poem titles and URLs
    poems = []
    for li in poems_list.find_all("li"):
        a_tag = li.find("a")
        if a_tag:
            poem_name = a_tag.text.strip()
            poem_url = a_tag["href"]
            poems.append((poem_name, poem_url))

    return poems

canon_poems = poems()
print("Canon Poems:", canon_poems)
```

```python
def poem_title_text(poemlist, n):
    if n not in [2, 5]:  # Only allow n = 2 or 5
        return None

    # Get the poem title and URL from the dictionary
    poem_title = list(poemlist.keys())[n]
    poem_url = poemlist[poem_title]

    # Fetch and parse the poem's webpage
    soup = get_soup(poem_url)

    # Locate the main content area of the poem
    content_div = soup.find("div", class_="mw-parser-output")

    # If content is not found, return an empty string
    if not content_div:
        return poem_title, ""

    # Extract text from all <p> elements inside the main content area
    paragraphs = content_div.find_all_next("p")[1:]  # Skip the first <p>
    poem_text = "\n".join([p.text.strip() for p in paragraphs])

    return poem_title, poem_text

# poems_data = poems()

# poem_title, poem_text = poem_title_text(poems_data, 5)
# print("Poem Title:", poem_title)
# print("Poem Text:", poem_text[:500], "...")
```
